```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/Prodigy internship/population.csv")
```

```python
data.head(200)
```

```python
data.describe
```

```python
#List of contries for the graph.
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France','India']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2023'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2023')
plt.title('Population of selected countries in 2023')
plt.xticks(rotation=45, ha='right') #Rotating the x-axis labels for readability
plt.show()
```
